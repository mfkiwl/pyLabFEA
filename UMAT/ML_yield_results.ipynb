````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image
size = (10,5)
plt.rcParams["figure.figsize"] = size
````

#### Data

````python
df1 = pd.read_csv('fsvc_data_test', engine='python', header=None, delimiter='\s+')
df2 = pd.read_csv('fsvc_data_1', engine='python', header=None, delimiter='\s+')
df3 = pd.read_csv('fsvc_data_01', engine='python', header=None, delimiter='\s+')
df4 = pd.read_csv('fsvc_data_001', engine='python', header=None, delimiter='\s+')
df5 = pd.read_csv('fsvc_data_biaxial_1', engine='python', header=None, delimiter='\s+')
df6 = pd.read_csv('fsvc_data_biaxial_01', engine='python', header=None, delimiter='\s+')
df7 = pd.read_csv('fsvc_data_biaxial_001', engine='python', header=None, delimiter='\s+')
````

### Uniaxial - X Loading

#### Different Increment sizes

````python
fsvc_1 = df2[1].to_numpy()
sigma_1 = df2[3].to_numpy()

fsvc_01 = df3[1].to_numpy()
sigma_01 = df3[3].to_numpy()

fsvc_001 = df4[1].to_numpy()
sigma_001 = df4[3].to_numpy()
````

````python
plt.scatter(sigma_1, fsvc_1, color='m')
plt.scatter(sigma_01, fsvc_01, color='b')
plt.scatter(sigma_001, fsvc_001, color='k')
plt.xlabel(r'$\sigma$ (MPa)')
plt.ylabel('fsvc')
plt.grid()
plt.show()
````

#### Error Margin - 10% of Equivalent stress
